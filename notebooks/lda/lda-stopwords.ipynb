This notebook is tools for finding and filtering stopwords

```python
import pandas as pd
from collections import defaultdict

import gensim.models

DATA_PATH = '/data/'
DOCS_PATH = '/data/events/lda/'

print('success')
```

```python
with open('stopwords_full.txt') as f:
    filter_words = set([word for line in f for word in line.split()])


def preprocessor(s: str):
    # remove empty words and words from filtered words 
    s = ' '.join(list(filter(lambda w: not (w in filter_words), s.split(' '))))
    return s


# load file and drop documents with empty text
df = pd.read_csv(DOCS_PATH + 'events_norm2.csv')

# preprocess of corpus
df['description'] = df['description'].apply(preprocessor)
df.to_csv(r'' + DOCS_PATH + 'events_norm2.csv', index=False)
print("success")
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
from gensim.models import Phrases

df = pd.read_csv(DOCS_PATH + 'events_norm2.csv').dropna()
docs = list(map(lambda s: s.split(), df['description'].tolist()))


bigram = Phrases(docs, min_count=5)

for idx in range(len(docs)):
    for token in bigram[docs[idx]]:
        if '_' in token:
            # Token is a bigram, add to document.
            docs[idx].append(token)
```

```python
from gensim.corpora import Dictionary

dictionary = Dictionary(docs)

# Filter out words that occur less than 20 documents, or more than 30% of the documents.
dictionary.filter_extremes(no_below=20, no_above=0.3)
# Bag-of-words representation of the documents.
corpus = [dictionary.doc2bow(doc) for doc in docs]
```

```python
from gensim.models import LdaMulticore

# Set training parameters.
num_topics = 40
chunksize = 11000
passes = 10
iterations = 100
eval_every = None  # Don't evaluate model perplexity, takes too much time.

# Make a index to word dictionary.
temp = dictionary[0]  # This is only to "load" the dictionary.
id2word = dictionary.id2token
```

```python
from gensim.models import LdaMulticore
lda = LdaMulticore(
    corpus=corpus,
    id2word=id2word,
    chunksize=chunksize,
    eta='auto',
    iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=eval_every
)
```
