```python
import gensim.models
import gensim

DATA_PATH = '/data/'
MODEL_PATH = DATA_PATH + 'models/wv1/'
model = gensim.models.Word2Vec.load(MODEL_PATH + 'mdl')
```

```python
with open('tests.txt') as f:
    tests = [[word for word in line.split()] for line in f]

for test in tests:
    print(model.wv.similarity(test[0], test[1]), test[0], test[1])
```

```python
import pandas as pd

df = pd.read_csv(DATA_PATH + 'events_norm.csv')
```

```python
ids = [24622, 24623, 24624, 24124, 24522, 24322]
words = []
colors = []
X = []

for i in range(len(ids)):
    for word in (df[df['id']==ids[i]]).iloc[0]['description'].split():
        try:
            X.append(model.wv[word])
            words.append(word)
            colors.append(str(ids[i])
            X.append(model.wv[word])
        except:
            print(word)
    

from sklearn.manifold import TSNE

tsne_model_en_2d = TSNE(perplexity=15, n_components=2, init='pca', n_iter=3500, random_state=32)
X_2d = tsne_model_en_2d.fit_transform(X)

print('success')
```

```python
import plotly.express as px


df_vis = pd.DataFrame(list(zip(*list(zip(*X_2d)), words, colors)), columns=['x', 'y', 'word', 'color'])

fig = px.scatter(df_vis, x="x", y="y", hover_name='word', color='color')
fig.show()
```
