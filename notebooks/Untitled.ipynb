```python
import numpy as np
from scipy.spatial import distance_matrix
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import pandas as pd
import plotly.express as px



X, y = make_blobs(n_samples=2000, n_features=2, centers=10, cluster_std=1.3)
df = pd.DataFrame(X, columns=['x', 'y'])
```

```python
from sklearn.metrics import davies_bouldin_score

scores = []
for n in range(2, 20):
    kmeans = KMeans(n_clusters=n, n_jobs=35).fit(X)
    scores.append([n, davies_bouldin_score(X, kmeans.labels_)])   
    
df_tmp = pd.DataFrame(scores, columns=['n_clusters', 'score'])
fig = px.line(df_tmp, x="n_clusters", y="score")
fig.show()
```

```python
def class_dist_max(a, b):
    max_dist = 0
    a_ind = clusters[a]
    b_ind = clusters[b]
    for i in a_ind:
        for j in b_ind:
            max_dist = max(max_dist, dists[i, j])
    return max_dist


def class_dist_min(a, b):
    min_dist = 1000000000
    a_ind = clusters[a]
    b_ind = clusters[b]
    for i in a_ind:
        for j in b_ind:
            min_dist = min(min_dist, dists[i, j])
    return min_dist


def class_dist_mean(a, b):
    sum_dists = 0
    a_ind = clusters[a]
    b_ind = clusters[b]
    edges = len(a_ind) * len(b_ind)
    for i in a_ind:
        for j in b_ind:
            sum_dists += dists[i, j]
            
    #print(a, b, len(a_ind), len(b_ind), sum_dists, sum_dists / len(a_ind), sum_dists / len(b_ind), sum_dists / edges)
    return sum_dists / edges**2
```
