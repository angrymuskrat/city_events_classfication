```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=1000, n_features=2, centers=10, cluster_std=0.7)
X_x, X_y = list(zip(*X))
```

```python
import pandas as pd
df = pd.DataFrame(zip(X_x, X_y, y), columns=['x', 'y', 'true_label'])
```

```python
import plotly.express as px
px.scatter(df, x='x', y='y', color='true_label').show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=2)
model = model.fit(X)
```

```python
df['label'] = model.labels_
import plotly.express as px
px.scatter(df, x='x', y='y', color='label').show()
```

```python
def get_elem(s: set):
    return next(iter(s))


n_samples = len(df)
clusters = dict([(i, set([i])) for i in range(n_samples)])
colors = list(range(n_samples))

for ind, pair in enumerate(model.children_):
    a, b = pair
    color_a, color_b = sorted([colors[get_elem(clusters[a])], colors[get_elem(clusters[b])]])
    clusters[ind + n_samples] = clusters[a] | clusters[b]
    colors = map(lambda c: c if c != color_b else color_a, colors)
    colors = list(map(lambda c: c - int(c > color_b), colors))
    if (n_samples - ind - 1) <= 50:
        df[n_samples - ind - 1] = colors
        
```

```python
px.scatter(df, x='x', y='y', color=10).show()
```
